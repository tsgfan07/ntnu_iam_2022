```python
def value_unit_converter(row, final_unit='kW'):
    options = ['kW', 'MW', 'GW']
    assert final_unit in options
    if final_unit=='kW':
        final_scaling=1
    elif final_unit=='MW':
        final_scaling=1/1000
    elif final_unit=='GW':
        final_scaling=1/1000000
    else:
        raise ValueError

    if 'GW' in row['unit']:
        row['value']  = row['value']*1000000*final_scaling
        row['unit'] = row['unit'].replace('GW', final_unit)
    elif 'MW' in row['unit']:
        row['value']  = row['value']*1000*final_scaling
        row['unit'] = row['unit'].replace('MW', final_unit)
    elif 'kW' in row['unit']:
        row['value']  = row['value']*final_scaling
        row['unit'] = row['unit'].replace('kW', final_unit)
    
    return row
```
